# Histograma en Python

En este curso, aprenderemos a crear un histograma utilizando el dataset del Titanic. El objetivo es entender la distribución de datos y cómo visualizarlos de manera efectiva.

### 1. Importar librerias e inspeccionar dataset

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.hist(
    "Age"
)
```

```python
df.hist(
    "Age",
    bins = 15
    
)
```

df.hist(
    "Age",

    bins = 15
    
)

### Histograma con seaborn

```python
import seaborn as sns
```

```python
sns.histplot(
    df,
    x ="Age"
)
```

El KDE (Kernel Density Estimate o Estimación de Densidad por Kernel) es una técnica para estimar la función de densidad de probabilidad de una variable aleatoria de manera suave, a partir de una muestra de datos. 

```python
sns.histplot(
    df,
    x ="Age",
    kde=True
)
```

```python
sns.histplot(
    df,
    x ="Age",
    kde=True,
    stat="probability"
)
```

```python
sns.histplot(
    df,
    x ="Age",
    kde=True,
    stat="probability",
    hue="Sex"
)
```
