* count: El número de valores no nulos en cada columna.

* std: La desviación estándar, que mide la dispersión de los datos.
* min: El valor mínimo de la columna.
* 25%: El primer cuartil, el valor que separa el 25% inferior de los datos.
* 50% (mediana): El segundo cuartil, el valor central de los datos.

* max: El valor máximo de la columna.

### Transfornacion de datos

```python
def mtoPies(estatura):
    resultado = estatura * 3.28084
    return resultado
```

```python
df["estatura en pies"] = df["estatura"].apply(mtoPies)
```
