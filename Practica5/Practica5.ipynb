```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df
```

```python
# Crear la gráfica de línea para temperatura máxima y mínima
plt.plot(df['dia'], df['temperatura_maxima'], label='Temperatura Máxima')
```

```python
# Crear la gráfica de línea para temperatura máxima y mínima
plt.plot(df['dia'], df['temperatura_maxima'], label='Temperatura Máxima', color='red', marker='o')

# Añadir títulos y etiquetas
plt.title('Temperaturas Máximas y Mínimas durante el Mes')
plt.xlabel('Día del Mes')
plt.ylabel('Temperatura (°C)')

# Añadir una leyenda
plt.legend()
```

### Multilinea

```python
# Crear la gráfica de línea para temperatura máxima y mínima
plt.plot(df['dia'], df['temperatura_maxima'], label='Temperatura Máxima', color='red', marker='o')
plt.plot(df['dia'], df['temperatura_minima'], label='Temperatura Mínima', color='blue', marker='o')

# Añadir títulos y etiquetas
plt.title('Temperaturas Máximas y Mínimas durante el Mes')
plt.xlabel('Día del Mes')
plt.ylabel('Temperatura (°C)')

# Añadir una leyenda
plt.legend()
```

```python
# Crear la gráfica de línea para temperatura máxima y mínima
plt.plot(df['dia'], df['temperatura_maxima'], label='Temperatura Máxima',  marker='o')
plt.plot(df['dia'], df['temperatura_minima'], label='Temperatura Mínima',  marker='o')

# Añadir títulos y etiquetas
plt.title('Temperaturas Máximas y Mínimas durante el Mes')
plt.xlabel('Día del Mes')
plt.ylabel('Temperatura (°C)')

# Añadir una leyenda
plt.legend()
```

```python
# Crear la gráfica de línea para temperatura máxima y mínima
plt.plot(df['dia'], df['temperatura_maxima'], label='Temperatura Máxima', color='red' #, marker='o')
        )

# Rellenar el área debajo de la línea de temperatura máxima
plt.fill_between(df['dia'], df['temperatura_maxima'], color='red', alpha=0.3)

# Añadir títulos y etiquetas
plt.title('Temperaturas Máximas y Mínimas durante el Mes')
plt.xlabel('Día del Mes')
plt.ylabel('Temperatura (°C)')

# Añadir una leyenda
plt.legend()
```
