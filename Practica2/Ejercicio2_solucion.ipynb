# Ejercicio: Realizar una grafica que muestre los alumnos de nuevo ingreso por cada carrera universitaria

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_escuela.csv")
df.head()
```

### Calcular columna

```python
conteo_carreras = df['Carrera'].value_counts() 
```

```python
print(conteo_carreras)
```

```python
plt.figure(figsize=(20, 8))  # Ajusta el tamaño de la figura según sea necesario
ax = conteo_carreras.plot(kind='bar', color=plt.get_cmap('tab10').colors)
ax.grid(True, linestyle='--', linewidth=0.5, color='grey')
ax.set_axisbelow(True)
plt.xticks(rotation=45, ha='right', fontsize=12) # Rotar las etiquetas del eje X
plt.title('Cantidad de alumnos nuevos 2024')
plt.xlabel('Carreras')
plt.ylabel('Cantidad alumnos')
plt.savefig('grafico_barras.png', format='png', dpi=300, bbox_inches='tight')
```
