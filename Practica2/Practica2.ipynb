# Grafico de barras en Python

En este curso, aprenderemos a crear gráficos de barras utilizando el dataset del Titanic. El objetivo es entender la distribución de datos y cómo visualizarlos de manera efectiva.

### 1. Importar librerias e inspeccionar dataset

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

### Analizar los datos para dar respuesta a la siguiente interrogante: ¿Cuantos pasajeros hombres y mujeres habian en el Titanic?

### 2. Funcion para hacer un conteo de datos

```python
conteo_sexo = df['Sex'].value_counts() 
```

```python
#El método value_counts() cuenta cuántas veces aparece cada valor único en la columna 
```

### 3. Gráfico de barras básica

```python
# plot() es un método de pandas que genera diferentes tipos de gráficos.
```

```python
# El parámetro kind es un argumento del método plot() que se utiliza para especificar el tipo de gráfico que deseas crear. 
```

```python
conteo_sexo.plot(kind='bar') 
```

### 4. Gráfico de barras horizontales

```python
conteo_sexo.plot(kind='barh')
```

### 5. Definir color al gráfico de barras.

```python
conteo_sexo.plot(kind='bar', color=['blue', 'red'])
```

```python
conteo_sexo.plot(kind='bar', color="red")
```

### 6. Definir ancho de barras

```python
conteo_sexo.plot(kind='bar', color="red", width = 0.9)
```

### 7. Definir titulos al gráfico de barras

```python
conteo_sexo.plot(kind='bar')
plt.title('Distribución de Género en el Titanic')
```

```python
conteo_sexo.plot(kind='bar')
plt.title('Distribución de Género en el Titanic')
plt.xlabel('Sexo de pasajero')
plt.ylabel('Cantidad de personas')
```

### 8. Establecer paleta de colores

```python
# Obtener la paleta de colores 'tab10' de Matplotlib
colors = plt.get_cmap('tab10').colors  

conteo_sexo.plot(kind='bar', color=colors)
plt.title('Distribución de Género en el Titanic')
plt.xlabel('Sexo de pasajeros')
plt.ylabel('Cantidad de personas')
```

### 9. Cuadricular grafica

```python
# Obtener la paleta de colores 'tab10' de Matplotlib
colors = plt.get_cmap('tab10').colors  

conteo_sexo.plot(kind='bar', color=colors)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey')
plt.title('Distribución de Género en el Titanic')
plt.xlabel('Sexo de pasajeros')
plt.ylabel('Cantidad de personas')


#plt.grid(True): Activa la cuadrícula en el gráfico.
#which='both': Añade la cuadrícula tanto en los ejes X como en los ejes Y.
#linestyle='--': Define el estilo de línea de la cuadrícula como punteado.
#linewidth=0.5: Establece el grosor de las líneas de la cuadrícula.
```

### 9.1 Cuadricula por atras de barras

ax es una instancia del objeto Axes, que es parte de un Figure en matplotlib. Se encarga de contener y manejar los datos gráficos, así como de realizar ajustes en las propiedades del gráfico.

El método ax.grid() en matplotlib se utiliza para agregar una cuadrícula a la gráfica. Esta cuadrícula ayuda a mejorar la legibilidad del gráfico al proporcionar líneas de referencia que facilitan la comparación de datos.

```python
# Obtener la paleta de colores 'tab10' de Matplotlib
colors = plt.get_cmap('tab10').colors  
ax = conteo_sexo.plot(kind='bar', color=colors)

ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey' )
ax.set_axisbelow(True)
plt.title('Distribución de Género en el Titanic')
plt.xlabel('Sexo de pasajeros')
plt.ylabel('Cantidad de personas')
```

### 12. Guardar gráfica

```python
# Obtener la paleta de colores 'tab10' de Matplotlib
colors = plt.get_cmap('tab10').colors  

conteo_sexo.plot(kind='bar', color=colors)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(rotation=0, ha='center', fontsize=10) # Rotar las etiquetas del eje X
plt.title('Distribución de Género en el Titanic',fontsize=14)
plt.xlabel('Sexo de pasajeros')
plt.ylabel('Cantidad de personas')
plt.savefig('grafico_barras.svg', 
            #format='png', 
            dpi=300, bbox_inches='tight')
```

+ PNG ('png'): Un formato de imagen de alta calidad y sin pérdida de calidad. Ideal para gráficos de alta resolución.  
+ JPG ('jpg' o 'jpeg'): Un formato con pérdida de calidad, adecuado para fotos y gráficos con menos detalles.  
+ PDF ('pdf'): Un formato vectorial que es ideal para impresión y escalado sin pérdida de calidad.  
+ SVG ('svg'): Un formato vectorial escalable, adecuado para gráficos que necesitan ser escalados a diferentes tamaños sin pérdida de calidad.

### 13. Agregar legend

```python
# Obtener la paleta de colores 'tab10' de Matplotlib
colors = plt.get_cmap('tab10').colors  # Obtener los colores del colormap 'tab10'

# Usar plt.bar para crear las barras
bars = plt.bar(conteo_sexo.index, conteo_sexo.values, color=colors)

# Agregar leyenda 
plt.legend(bars, conteo_sexo.index, title='Género', loc='upper right')

# Personalizar el gráfico
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(rotation=0, ha='center', fontsize=10)  # Rotar las etiquetas del eje X
plt.title('Distribución de Género en el Titanic', fontsize=14)
plt.xlabel('Sexo de pasajeros')
plt.ylabel('Cantidad de personas')
```

### 14. Agregar valores a cada barra

```python
# Obtener la paleta de colores 'tab10' de Matplotlib
colors = plt.get_cmap('tab10').colors  # Obtener los colores del colormap 'tab10'
# Usar plt.bar para crear las barras
bars = plt.bar(conteo_sexo.index, conteo_sexo.values, color=colors)
#---------------------------------------------------------------------
# Agregar los valores encima de cada barra
for bar in bars:
    yval = bar.get_height()  # Obtener la altura de cada barra (el valor)
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Posición en x (centro de la barra)
        yval,  # Posición en y (altura de la barra)
        int(yval),  # El valor que se va a mostrar
        ha='center',  # Alinear horizontalmente en el centro
        va='bottom',  # Alinear verticalmente justo encima de la barra
        fontsize=10  # Tamaño de la fuente
    )

#---------------------------------------------------------------------
# Agregar leyenda 
plt.legend(bars, conteo_sexo.index, title='Género', loc='upper right')
# Personalizar el gráfico
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(rotation=0, ha='center', fontsize=10)  # Rotar las etiquetas del eje X
plt.title('Distribución de Género en el Titanic', fontsize=14)
plt.xlabel('Sexo de pasajeros')
plt.ylabel('Cantidad de personas')
```
