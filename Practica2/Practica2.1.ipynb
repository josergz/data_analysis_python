```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bd_flood_2000_2024.csv")
df.head()
```

```python
eventos_por_aÃ±o = df['Start Year'].value_counts().sort_index()

#sort_index(): Ordena la Serie resultante por su Ã­ndice, es decir, ordena los aÃ±os en orden ascendente (cronolÃ³gicamente).
```

```python
bars = plt.bar(eventos_por_aÃ±o.index, eventos_por_aÃ±o.values)
```

```python
bars = plt.bar(eventos_por_aÃ±o.index, eventos_por_aÃ±o.values)
plt.xticks(eventos_por_aÃ±o.index, rotation='vertical', fontsize=8)
```

```python
import numpy as np
colores = plt.cm.viridis(np.linspace(0, 1, len(eventos_por_aÃ±o)))
bars = plt.bar(eventos_por_aÃ±o.index, eventos_por_aÃ±o.values, color=colores)
plt.xticks(eventos_por_aÃ±o.index, rotation='vertical', fontsize=8)
plt.title(f'Inundaciones en el mundo de {eventos_por_aÃ±o.index.min()} a {eventos_por_aÃ±o.index.max()}', fontsize=14)
plt.xlabel('AÃ±o')
plt.ylabel('NÃºmero de inundaciones')
```

```python
import numpy as np
colores = plt.cm.viridis(np.linspace(0, 1, len(eventos_por_aÃ±o)))
bars = plt.bar(eventos_por_aÃ±o.index, eventos_por_aÃ±o.values, color=colores)
plt.xticks(eventos_por_aÃ±o.index, rotation='vertical', fontsize=8)
plt.title(f'Inundaciones en el mundo de {eventos_por_aÃ±o.index.min()} a {eventos_por_aÃ±o.index.max()}', fontsize=14)
plt.xlabel('AÃ±o')
plt.ylabel('NÃºmero de inundaciones')
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey')
```

```python
colores = plt.cm.viridis(np.linspace(0, 1, len(eventos_por_aÃ±o)))
bars = plt.bar(eventos_por_aÃ±o.index, eventos_por_aÃ±o.values, color=colores)
#---------------------------------------------------------------------
# Agregar los valores encima de cada barra
for bar in bars:
    yval = bar.get_height()  # Obtener la altura de cada barra (el valor)
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # PosiciÃ³n en x (centro de la barra)
        yval,  # PosiciÃ³n en y (altura de la barra)
        int(yval),  # El valor que se va a mostrar
        ha='center',  # Alinear horizontalmente en el centro
        va='bottom',  # Alinear verticalmente justo encima de la barra
        fontsize=10  # TamaÃ±o de la fuente
    )
#---------------------------------------------------------------------
plt.xticks(eventos_por_aÃ±o.index, rotation='vertical', fontsize=8)
plt.title(f'Inundaciones en el mundo de {eventos_por_aÃ±o.index.min()} a {eventos_por_aÃ±o.index.max()}', fontsize=14)
plt.xlabel('AÃ±o')
plt.ylabel('NÃºmero de inundaciones')
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey')
```
